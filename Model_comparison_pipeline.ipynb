# Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

# Import data

```python
dataset = pd.read_csv('preprocessed_data.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
# Import ependent Variable
X = dataset.iloc[:, 0:18].values
```

```python
# Import Dependent Variable
#y = dataset.iloc[0:100000, 29].values
y = dataset.iloc[0:, 19].values
```

# Split data and feature scaling

```python
#Split data into training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Fit model to training set

```python
#execution time estimation
start = time.time() 

#import pipeline
from sklearn.pipeline import Pipeline

# Fitting Kernel SVM to the Training set
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression


classifiers = [
    SVC(kernel = 'rbf',probability=True),
    RandomForestClassifier(n_estimators = 200, criterion = 'entropy'),
    GaussianNB(),
    LogisticRegression()
    ]


for classifier in classifiers:
    pipe = Pipeline(steps=[('classifier', classifier)])
    pipe.fit(X_train, y_train)   
    print(classifier)
    print("model score: %.3f" % pipe.score(X_test, y_test))

#execution time estimation
end = time.time()
print(end - start)
```

```python
# Applying k-Fold Cross Validation
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score

#execution time estimation
start = time.time()

for classifier in classifiers:
    pipe = Pipeline(steps=[('classifier', classifier)])
    accuracies = cross_val_score(pipe, X = X_train, y = y_train, cv = 10)  
    print(classifier)
    print("model score: %.3f" % accuracies.mean())



#accuracies.std()


#execution time estimation
end = time.time()
print(end - start)
```

# Perdict test set and evaluate the model

```python
# Predicting the Test set results
# Making the Confusion Matrix

from sklearn.metrics import classification_report
for classifier in classifiers:
    pipe = Pipeline(steps=[('classifier', classifier)])
    y_pred = pipe.predict(X_test)  
    class_report = classification_report( y_test, y_pred)
    print(classifier)
    print(class_report)
```

```python
#evaluate test result accuracy and so on
from sklearn.metrics import accuracy_score

for classifier in classifiers:
    pipe = Pipeline(steps=[('classifier', classifier)])
    y_pred = pipe.predict(X_test)  
    accuracy = accuracy_score(y_test, y_pred)
    print(classifier)
    print ("accuracy:",accuracy)
```
